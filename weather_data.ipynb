WEATHER DATASET:
1. DATE/TIME : 1/1/2012   0:00  The data was recorded on January 1st, 2012, at midnight(00:00)
2. Temperature in Celcius : -1.8  Thr air temp at that time. negative means below freezing
3. Dew point temperture (*C) : The temperature at which air becomes saturated and dew forms- lower values means drier air.
4. Relative Humidity(%) : Shows how much moisture is in the air (0% = dry, 100% = fully saturated)
5. Wind Speed in kilometers per hour : How fast womd is blowing
6. Visibility in kilometers : How far ahead you can see- lower values means foggy or misty conditions
7. Atmospheric pressure in kilopascals  : Air pressure- usually around 101 kpa at sea level. Changes indicated weather shifts.
Pressure is measured in kilopascals (kPa).
1 kilopascal = 1000 pascals (a unit of pressure).
At sea level, normal atmospheric pressure is around 101 kPa — that’s why in our dataset, values like 101.24 are totally normal.
Pressure Trend	                        What It Usually Means
High pressure (e.g., 102–104 kPa):	  Clear, sunny, and calm weather
Low pressure (e.g., 98–100 kPa):  	  Cloudy, windy, or rainy weather
Rapidly dropping pressure	           Storm or rain might be coming
Rising pressure	                     Weather is getting clearer
Press_kPa = 101.24
That’s just slightly above the average — so it means stable and calm weather (which matches your “Fog” condition, no stormy signs).
8. Type of weather conditions observed : type of weather -> fog, freezing, drizzle

Standard deviation (SD) is a number that tells you how spread out or scattered the values in a dataset are.

Small SD → values are close to the average

Large SD → values are far from the average, more spread out

```python
standard_deviation=data['Press_kPa'].std()
print("Standard deviation (SD) is a number that tells \n you how spread out or scattered the values in a dataset are.\nSmall SD → values are close to the average\nLarge SD → values are far from the average, more spread out")
print("\nSTANDARD DEVIATION of Pressure is: ",standard_deviation)
#std= root(∑(xi​−xˉ)2​ /N -1)
#where xi = each value, xˉ = means,n = number of values
```

Variance measures how spread out the values in a dataset are from their mean (average).

Small variance → values are close to the mean → very consistent

Large variance → values are far from the mean → more spread out

```python
print("Variance measures how spread out the values in\n a dataset are from their mean (average).\nSmall variance → values are close to the mean → very consistent\n Large variance → values are far from the mean → more spread out")

print("\nVariance of Relative Humidity is:",data['Rel Hum_%'].var())
```
